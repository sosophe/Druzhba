```python
import pandas as pd
pd.options.mode.chained_assignment = None 
import numpy as np
import re
import nltk

from gensim.models import word2vec

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def tsne_plot(data):
    "Creates and TSNE model and plots it"
#     sample = data.values()
    labels = list(data.keys())
    X = np.array(np.stack(data.values(), axis=0))
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    array = np.array(X, dtype=float)

    new_values = tsne_model.fit_transform(X)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
#                      xytext=(5, 2),
#                      textcoords='offset points',
#                      ha='right',
#                      va='bottom')
                )
    plt.show()
```

```python
tsne_plot(df_cats)
```

```python
def tsne_plot_nolabels(data):
    "Creates and TSNE model and plots it"
#     sample = data.values()
    labels = list(data.keys())
    X = np.array(np.stack(data.values(), axis=0))
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    array = np.array(X, dtype=float)

    new_values = tsne_model.fit_transform(X)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    trace0 = go.Scatter(
        x = x,
        y = y,
        mode = 'markers',
        showlegend = False,
    #     text=most_similar_words_preprocessed_sentences, #TODO: change this
        marker = dict(
            size = 8,
            colorscale ='Jet',
            showscale = False,
            line = dict(
                width = 2,
                color = 'rgb(255, 255, 255)'
            ),
            opacity = 0.8
        )
    )
    data = [trace0]

    layout = dict(title = 'TSNE Exploration',
                  hovermode= 'closest',
                  yaxis = dict(zeroline = False),
                  xaxis = dict(zeroline = False),
                  showlegend= True
                 )

    fig = dict(data=data, layout=layout)
    py.iplot(fig, filename='styled-scatter')
```

```python
from sklearn.cluster import KMeans


X = np.array(np.stack(df_cats.values(), axis=0))
wcss=[]
for i in range(1,20):
    kmeans=KMeans(n_clusters=i,init='k-means++',)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,20),wcss)
plt.title('Elbow Method')
plt.xlabel('No. of cluster')
plt.ylabel('wcss: sum of dist. of sample to their closest cluster center' )
```
