```python
df_ex = df_ex.merge(df_cat, left_on=['ИД Каталожной записи'], right_on='doc_id')
```

```python
df_vyd = pd.read_csv("./Датасеты/Vydacha_1.csv")
```

```python
df_vyd = df_vyd.append(pd.read_csv("./Датасеты/Vydacha_2.csv"))
```

```python
df_vyd.shape
```

```python
df_vyd.head()
```

```python
df_vyd = df_vyd.merge(df_ex, on='Штрих-код')
```

```python
df_vyd.shape
```

```python
df_vyd.head()
```

```python
df_vyd['rating'] = 1
```

```python
df_vyd.columns
```

```python
from tqdm import tqdm
```

```python
def update_weight(group):
    weights = group['p650a'].value_counts(normalize=True).to_dict()
    group['rating'] = group['p650a'].map(weights) 
    return group
```

```python
tqdm.pandas()
```

```python
df_vyd = df_vyd.groupby('ИД читателя').progress_apply(update_weight)
```
